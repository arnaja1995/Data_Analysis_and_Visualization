```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
# Using Joint Plot
sns.jointplot(x='fare', y='age', data=titanic, kind='hex')
```

```python
# Using Distribution Plot
sns.displot(titanic['fare'],bins=28,kde=False,color='red')
```

```python
# Using Box Plot (Categorical Plot)
sns.boxplot(x='class',y='age',data=titanic,palette='coolwarm')
```

```python
# Using Swarm Plot (Categorical Plot)
sns.swarmplot(x='class',y='age',data=titanic,palette='Set2')
```

```python
# Using Violin Plot (Categorical Plot)
sns.violinplot(x='class',y='age',data=titanic,palette='Set1')
```

```python
# Using Heatmap (Matrix Plot)
ti=titanic.corr()
sns.heatmap(ti,cmap='Purples')
```

```python
# Customizing PairGrid by using Facet Grid 
Tt=sns.FacetGrid(data=titanic,col='sex',row='class')
Tt.map(plt.hist,'age',color='green')
```
